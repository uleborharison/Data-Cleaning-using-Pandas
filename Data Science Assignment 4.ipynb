<h2>Data Cleaning using Pandas(Python Library)</h2>

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\Huleb\OneDrive\Desktop\Data Science Program Files\Data Sets\messy_data.csv")
#importing the csv file for cleaning
```

```python
data.duplicated().sum()  #Check for duplicates using TransactionID column as the primary key.
```

```python
data.drop_duplicates(inplace = True) #drop all duplicates that is in 'TrnsnID' Column
```

```python
data.isna().sum() #checking for missing values
```

```python
#with specific, I will be filling each column with specific members and not generalizing the fil.
data['ProdtName'].bfill(inplace = True) 
data['PaymentType'].bfill(inplace = True)
```

```python
data.isna().sum()
```

```python
data['Data'] = pd.to_datetime(data['Data'], format ='mixed') #this  gives the Date column members an appropraite date format
```

```python
data['PaymentType'].replace({'debit card':'Debit Card','Debit':'Debit Card','Credit':'Credit Card', 'paypal':'Paypal', 'cash': 'Cash'}, inplace = True)
#This renames the PaymentType column members' value appropriately
```

```python
data['CusterAge']= np.abs(data['CusterAge']) # this is use to remove the negative signs attached to a customerAge
```

```python
data['Qutity'] = np.abs(data['Qutity'])
```

```python
#i want to add this here, because it can be two ways to salve the zero in quantity, is either we use the mean of quantity or we just ignore and use the value zero(0), but here i am adding the mean to zero values of the cell. It also takes out unrealistic values
data['Qutity'] = data['Qutity'].mask((data['Qutity'] < 1),data['Qutity'].mean()).astype(int)
```

```python
data['ProdtName'].replace({'Boo':'Book','Bok':'Book', 'Bookk':'Book', }, regex = True, inplace = True)
data['ProdtName'].replace('kk', 'k', regex=True, inplace = True)
#Worked on renaming the productName column members value appropriately
```

```python
#this is used to make the column TotalSales have a valid calculation of the sales made, which TotalSale = Quantity * by Price
data['TtlSales'] = data['TtlSales'].mask(data['TtlSales'] != (np.multiply(data['Prize'] , data['Qutity'])), np.multiply(data['Prize'], data['Qutity']), axis = 0).astype(float)
```

```python
data.rename(columns = {'TrnsnID':'TransactionID','CusmrID':'CustomerID','ProdtID':'ProductID','ProdtName':'ProductName', 'CusterAge': 'CustomerAge', 'TtlSales':'TotalSales', 'Qutity':'Quantity','Prize':'Price','Data':'Date'}, inplace = True)
#This renames the column titles to the appropriate spelling
```

```python
#Sort the date to get the actual dates of transactions, so the transactionID is our primary key, since the values must be unique without repetition
data.sort_values(by=['Date'], inplace=True)  

#This is indexed for easy retrieval of data and takes off the Unrealistic Values
data.set_index('TransactionID', inplace = True) 
```
